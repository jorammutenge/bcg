```python
import pandas as pd
import numpy as np
import os
```

```python
# Read in data as batch
folder_path = 'data'
df = [pd.read_csv(os.path.join(folder_path, filename), parse_dates=['Date'], index_col=0)
              for filename in os.listdir(folder_path) if filename.endswith('.csv')]
```

```python
# get company names based on csv files names
from file_names import get_csv_filenames
companies = [company for company in get_csv_filenames(folder_path)]
modified_companies = ', '.join(companies)
modified_companies
```

```python
# assign company name to corresponding dataframe
AAPL, AMZN, BYDDY, GOOGL, JNJ, MRNA, MSFT, PFE, SIEGY, TSLA = df[:10]
```

```python
dfs = [AAPL, AMZN, BYDDY, GOOGL, JNJ, MRNA, MSFT, PFE, SIEGY, TSLA]
```

```python
# add new calculated columns
modified_dfs = []

for df in dfs:
    modified_df = (
        df.assign(
            MA50=lambda df_: df_.Close.rolling(50).mean(),
            MA200=lambda df_: df_.Close.rolling(200).mean(),
            Prev_day_close=lambda df_: df_.Close.shift(1),
            Change_in_price=lambda df_: df_.Close - df_.Prev_day_close,
            Pct_change=lambda df_: df_.Close.pct_change(),
            Prev_day_vol=lambda df_: df_.Volume.shift(1),
            Change_in_vol=lambda df_: df_.Volume - df_.Prev_day_vol,
            Pct_change_in_vol=lambda df_: df_.Volume.pct_change()
        )
    )
    modified_dfs.append(modified_df)
    
AAPL, AMZN, BYDDY, GOOGL, JNJ, MRNA, MSFT, PFE, SIEGY, TSLA = modified_dfs[:10]
```

```python
# write to csv
dfs_ = [AAPL, AMZN, BYDDY, GOOGL, JNJ, MRNA, MSFT, PFE, SIEGY, TSLA]

names = ['AAPL', 'AMZN', 'BYDDY', 'GOOGL', 'JNJ', 'MRNA', 'MSFT', 'PFE', 'SIEGY', 'TSLA']

for df, name in zip(dfs_, names):
    filename = f'data/{name}.csv'  # Generate the filename based on the name
    df.to_csv(filename, index=True)
```

### Candlestick plot

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates

fig, ax = plt.subplots(figsize=(10,5), dpi=200)

def plot_candle(df, ax):
    # wick
    ax.vlines(x=df.index, ymin=df.Low, ymax=df.High, colors='k', linewidth=1)
    # red - decrease
    red = df.query('Open > Close')
    ax.vlines(x=red.index, ymin=red.Close, ymax=red.Open, colors='r', linewidth=3)
    # green - increase
    green = df.query('Open < Close')
    ax.vlines(x=green.index, ymin=green.Close, ymax=green.Open, colors='g', linewidth=3)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return df


(AAPL
 .resample('d')
 .agg({'Open':'first', 'High':'max', 'Low':'min', 'Close':'last'})
 .loc['jan 2022': 'jun 2022']
 .pipe(plot_candle, ax)
)
```
